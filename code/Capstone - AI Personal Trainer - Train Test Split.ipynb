```python
#features
X = workout_df.drop('body class', axis=1)

#target value
y = workout_df['body class']
```

```python
#create a train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Train Machine Learning Classification Model

```python
#import and build a ML pipepline
from sklearn.pipeline import make_pipeline

#standardizes the data
from sklearn.preprocessing import StandardScaler

#import classifiers
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
#setup the separate ML for each classifier thru pipelines
#go through 4 different algorithms to find the best score and use that as the classifier
pipelines = {
    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'rc':make_pipeline(StandardScaler(), RidgeClassifier()),
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gb':make_pipeline(StandardScaler(), GradientBoostingClassifier())
}
```

```python
#start with a empty dictionary
fit_models = {}

#create a for loop to go through each classifier under pipeline
for algorithm, pipeline in pipelines.items():
    model = pipeline.fit(X_train, y_train)
    fit_models[algorithm] = model
```

```python
fit_models
```

## Evaluate and Serialize Model

```python
from sklearn.metrics import accuracy_score
import pickle
```

```python
#create a for loop to run the x_test
for algorithm, model in fit_models.items():
    #create a variable to go thru each classifier to predict the algorithm
    yhat = model.predict(X_test)
    #print out the accuracy score for each classifier
    print(algorithm, accuracy_score(y_test, yhat))
```

Gradient Boosting Classifier seems to give the best result so I will pickle the gb out

```python
#export the gb model as a pickle file
with open('../Capstone/data/workout_classifier.pkl', 'wb') as f:
    pickle.dump(fit_models['gb'],f)
```
